Retrieve the timezones and generate a dataframe with the timezones and the amount of airports in them. Plot it as a histogram as well.

```python
query="""SELECT timezone, count(name) AS Amount_of_Airport 
FROM airports GROUP BY timezone ORDER BY Amount_of_Airport ASC;"""
df = pd.read_sql_query(query, connection)
df.head()
```

```python
df1=df.set_index('timezone')
```

```python
df1.plot.hist()
```

Which country has more airlines? And which one has less?

```python
query="""SELECT country, count(active) AS NUM_OF_AIRPORTS 
FROM airlines GROUP BY country ORDER BY NUM_OF_AIRPORTS DESC;"""
df = pd.read_sql_query(query, connection)
df.head()
```

```python
df.loc[df['NUM_OF_AIRPORTS'].idxmax()]
```

```python
df.loc[df['NUM_OF_AIRPORTS'].idxmin()]
```

Which city has the most outgoing destinations? And the least?

```python
query="""SELECT CAST(airports.city AS TEXT) AS Source_CITY , count(source) as freq
FROM routes
INNER JOIN airports
ON routes.source=airports.code
GROUP BY Source_CITY ORDER BY freq desc;"""
df = pd.read_sql_query(query, connection)
df.head()
```

```python
df.loc[df['freq'].idxmax()]
```

```python
df.loc[df['freq'].idxmin()]
```

And which city has the most incoming destinations? And the least?

```python
query="""SELECT CAST(airports.city AS TEXT) AS Dest_CITY , count(source) as freq
FROM routes
INNER JOIN airports
	ON routes.dest=airports.code
GROUP BY Dest_CITY ORDER BY freq desc;"""
df = pd.read_sql_query(query, connection)
df.head()
```

```python
df.loc[df['freq'].idxmax()]
```

```python
df.loc[df['freq'].idxmin()]
```
