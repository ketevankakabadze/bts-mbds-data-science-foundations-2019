```python
sorted_data = data.groupby([(data.index.year), (data.index.month)]).mean()
sorted_data.head()
```

```python
sorted_data.index.names=['Year','Month']
sorted_data.head()
```

```python
sorted_by_highest_month=sorted_data.loc[sorted_data.groupby('Year')['Total'].idxmax()]
sorted_by_highest_month
```

Plot each year with the number of bikes to also see it visually.
